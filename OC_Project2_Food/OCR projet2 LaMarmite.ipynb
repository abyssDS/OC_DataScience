```python
#Get some data with always filled columns to work on
crucial_columns = raw_data[['code'
                 ,'created_datetime'
                 ,'last_modified_datetime'
                 ,'product_name'
                 ,'nutrition_grade_fr'
                 ,'nutrition-score-fr_100g'
                 ,'nutrition-score-uk_100g'
                 ,'additives_n'
                 ,'fat_100g'
                 ,'energy_100g'
                 ,'trans-fat_100g'
                 ,'cholesterol_100g'
                 ,'saturated-fat_100g'
                 ,'carbohydrates_100g'
                 ,'sugars_100g'
                 ,'fiber_100g'
                 ,'proteins_100g'
                 ,'salt_100g'
]]

data = crucial_columns.dropna(how='any')
#keep only rows where all above features are known
```

```python
data.shape #130885
data.
```

```python
#Remove some bad values (like above 100% or less than 0%)
for column in data:
    if '-' in column:
        data=data.rename(columns = {column:(column.replace('-', '_'))})
for column in data:
    if '100g' in column  and column!='energy_100g' and 'nutrition' not in column:
        print('Removing bad values for : '+column)
        data = data.query(column+' <= 100 & '+column+'>=0')
        
data.shape
```
